```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
UCI-HAR Decision Tree pipeline (raw signals → TSFEL-like features)
— Trains on UCI-HAR raw accelerometer signals
— Evaluates on your collected smartphone trials

Folders expected:
UCI_HAR_ROOT/
  train/
    Inertial Signals/
      body_acc_x_train.txt
      body_acc_y_train.txt
      body_acc_z_train.txt
      total_acc_x_train.txt
      total_acc_y_train.txt
      total_acc_z_train.txt
    y_train.txt
  test/
    Inertial Signals/ (same 6 files)
    y_test.txt

USER_DATA_ROOT/
  Downstarirs/*.csv         (note: spelled as in your zip)
  Upstairs/*.csv
  Sitting/*.csv
  Standing/*.csv
  Laying/*.csv

Each CSV has columns like: ["time", "ax (m/s^2)", "ay (m/s^2)", "az (m/s^2)", "aT (m/s^2)"]

Outputs:
- metrics printed (accuracy, precision, recall)
- confusion matrix figure: ./confusion_matrix_user.png
"""

import os
import glob
import math
import numpy as np
import pandas as pd
from pathlib import Path
from typing import Tuple, Dict, List
from scipy.stats import entropy
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from itertools import product

# ============================
# USER: SET THESE TWO PATHS
# ============================
UCI_HAR_ROOT = r"C:\Users\RAJDEEP VRAJ\OneDrive - iitgn.ac.in\Semester 7\Machine learning\es335-25-fall-assignment-1-master\HAR\human+activity+recognition+using+smartphones\UCI HAR Dataset\UCI HAR Dataset"
USER_DATA_ROOT = r"C:\Users\RAJDEEP VRAJ\OneDrive - iitgn.ac.in\Semester 7\Machine learning\es335-25-fall-assignment-1-master\HAR_task_3"

# ============================
# CONFIG (aligned with UCI-HAR)
# ============================
TARGET_FS = 50           # 50 Hz (UCI-HAR uses 50 Hz)
WIN_SEC = 2.56           # 2.56 s window
WIN = int(WIN_SEC * TARGET_FS)  # 128 samples
STEP = WIN // 2          # 50% overlap = 64
FIRST_N_SECONDS = 10     # keep first 10 seconds only
RAW_COLS = ["ax", "ay", "az"]

# Mapping your folder names -> UCI-HAR integer labels
# UCI-HAR labels: 1 WALKING, 2 WALKING_UPSTAIRS, 3 WALKING_DOWNSTAIRS, 4 SITTING, 5 STANDING, 6 LAYING
FOLDER_TO_LABEL = {
    "Upstairs": 2,
    "Downstarirs": 3,  
    "Sitting": 4,
    "Standing": 5,
    "Laying": 6,
}
LABEL_TO_NAME = {
    1: "WALKING",
    2: "WALKING_UPSTAIRS",
    3: "WALKING_DOWNSTAIRS",
    4: "SITTING",
    5: "STANDING",
    6: "LAYING",
}

# ----------------------------
# Utility: simple, robust features 
# ----------------------------
def feat_time_series(x: np.ndarray) -> Dict[str, float]:
    x = np.asarray(x)
    if x.size == 0:
        return {
            "mean": np.nan, "std": np.nan, "median": np.nan, "mad": np.nan,
            "min": np.nan, "max": np.nan, "iqr": np.nan, "energy": np.nan,
            "zcr": np.nan, "mcr": np.nan, "skew": np.nan, "kurt": np.nan,
        }
    mean = np.mean(x)
    std = np.std(x, ddof=1) if x.size > 1 else 0.0
    median = np.median(x)
    mad = np.median(np.abs(x - median))
    x_min = np.min(x)
    x_max = np.max(x)
    q75, q25 = np.percentile(x, [75, 25])    # split the data into 75% and #25% help us to detect the outlier by calculuate iqr
    iqr = q75 - q25
    energy = np.sum(x**2) / x.size
    # zero crossing rate ( no of sample changing the sign means we can capture the oscillation)
    zcr = np.mean(np.diff(np.sign(x)) != 0)
    # mean crossing rate ( to determine small or high variation in acceleration)
    mcr = np.mean((x[:-1] - mean) * (x[1:] - mean) < 0) if x.size > 1 else 0.0
    # moments ( to descirbe spread and dispersion around the mean)
    if std > 1e-12:
        skew = np.mean(((x - mean) / std) ** 3)                       # asymmetry
        kurt = np.mean(((x - mean) / std) ** 4) - 3.0                 
    else:
        skew, kurt = 0.0, 0.0
    return {
        "mean": mean, "std": std, "median": median, "mad": mad,
        "min": x_min, "max": x_max, "iqr": iqr, "energy": energy,
        "zcr": zcr, "mcr": mcr, "skew": skew, "kurt": kurt,
    }

def feat_freq_series(x: np.ndarray, fs: int) -> Dict[str, float]:
    # one-sided FFT, ignore DC-only tiny windows robustly
    x = np.asarray(x)
    if x.size == 0:
        return {"spec_centroid": np.nan, "spec_entropy": np.nan, "dom_freq": np.nan, "dom_amp": np.nan}
    X = np.fft.rfft(x * np.hanning(x.size))
    mag = np.abs(X)
    freqs = np.fft.rfftfreq(x.size, d=1.0/fs)
    p = mag**2
    psum = np.sum(p)
    if psum <= 1e-12:
        return {"spec_centroid": 0.0, "spec_entropy": 0.0, "dom_freq": 0.0, "dom_amp": 0.0}
    p_norm = p / psum
    spec_centroid = np.sum(freqs * p_norm)
    spec_entropy = entropy(p_norm + 1e-12, base=2)
    idx = int(np.argmax(mag))
    dom_freq = float(freqs[idx])
    dom_amp = float(mag[idx])
    return {"spec_centroid": spec_centroid, "spec_entropy": spec_entropy, "dom_freq": dom_freq, "dom_amp": dom_amp}

def pairwise_corr(a: np.ndarray, b: np.ndarray) -> float:
    if a.size < 2 or b.size < 2:
        return 0.0
    if np.std(a) < 1e-12 or np.std(b) < 1e-12:
        return 0.0
    return float(np.corrcoef(a, b)[0,1])

def extract_feature_vector(ax: np.ndarray, ay: np.ndarray, az: np.ndarray, fs: int) -> Dict[str, float]:
    # Per-axis stats
    fx = feat_time_series(ax); fy = feat_time_series(ay); fz = feat_time_series(az)
    Fx = feat_freq_series(ax, fs); Fy = feat_freq_series(ay, fs); Fz = feat_freq_series(az, fs)
    # magnitude channel
    m = np.sqrt(ax**2 + ay**2 + az**2)
    fm = feat_time_series(m); Fm = feat_freq_series(m, fs)
    # correlations
    c_xy = pairwise_corr(ax, ay)
    c_xz = pairwise_corr(ax, az)
    c_yz = pairwise_corr(ay, az)

    # flatten with consistent keys
    feats = {}
    for axis, d in zip(["x","y","z","mag"], [fx, fy, fz, fm]):
        for k, v in d.items():
            feats[f"{axis}_{k}"] = v
    for axis, d in zip(["x","y","z","mag"], [Fx, Fy, Fz, Fm]):
        for k, v in d.items():
            feats[f"{axis}_{k}"] = v
    feats["corr_xy"] = c_xy
    feats["corr_xz"] = c_xz
    feats["corr_yz"] = c_yz
    return feats

# ----------------------------
# Windowing
# ----------------------------
def sliding_windows(n: int, win: int, step: int):
    start = 0
    while start + win <= n:
        yield start, start + win
        start += step

# ----------------------------
# Load UCI-HAR raw (total acceleration is fine)
# ----------------------------
def load_uci_split(root: str, split: str) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    sig_dir = Path(root) / split / "Inertial Signals"
    def load_sig(basename: str):
        path = sig_dir / basename
        # each file: samples x 128 (windows) -> shaped (n_windows, 128)
        arr = np.loadtxt(path)
        return arr  # shape: (num_windows, 128)

    # choose total acceleration to be closest to your phone raw (gravity + body)
    ax = load_sig(f"total_acc_x_{split}.txt")
    ay = load_sig(f"total_acc_y_{split}.txt")
    az = load_sig(f"total_acc_z_{split}.txt")
    # labels (per window)
    y = np.loadtxt(Path(root) / split / f"y_{split}.txt").astype(int)
    return ax, ay, az, y

def featurise_uci(ax_2d: np.ndarray, ay_2d: np.ndarray, az_2d: np.ndarray, fs: int) -> pd.DataFrame:
    rows = []
    for i in range(ax_2d.shape[0]):
        feats = extract_feature_vector(ax_2d[i], ay_2d[i], az_2d[i], fs)
        rows.append(feats)
    return pd.DataFrame(rows)

# ----------------------------
# Load & process YOUR data
# ----------------------------
def read_user_csv(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    # unify column names
    colmap = {
        "ax (m/s^2)": "ax",
        "ay (m/s^2)": "ay",
        "az (m/s^2)": "az",
        "aT (m/s^2)": "aT",
        "time": "time",
    }
    df = df.rename(columns=colmap)
    # guard rails
    for c in ["ax","ay","az"]:
        if c not in df.columns:
            raise ValueError(f"Missing column {c} in {path}")
    if "time" not in df.columns:
        # create synthetic time index @100 Hz if absent
        df["time"] = np.arange(len(df)) / 100.0
    return df[["time","ax","ay","az"]].copy()

def trim_and_downsample(df: pd.DataFrame, assume_fs: int = 100, target_fs: int = 50, seconds: int = 10) -> pd.DataFrame:
    # keep first N seconds by time
    df_10 = df[df["time"] <= seconds].copy()
    if df_10.empty:
        # fallback: take first seconds * assume_fs rows
        df_10 = df.iloc[:seconds * assume_fs].copy()
        df_10["time"] = np.arange(len(df_10)) / assume_fs

    # downsample 100Hz -> 50Hz by taking every 2nd sample (robust & simple)
    factor = int(assume_fs // target_fs)
    if factor < 1:
        raise ValueError("assume_fs must be >= target_fs")
    df_50 = df_10.iloc[::factor].reset_index(drop=True)

    # reindex time to exact 50Hz grid (optional)
    df_50["time"] = np.arange(len(df_50)) / target_fs
    # ensure exactly seconds * target_fs rows if possible
    need = seconds * target_fs
    if len(df_50) >= need:
        df_50 = df_50.iloc[:need].copy()
    return df_50

def window_and_featurise_user(df: pd.DataFrame, fs: int) -> Tuple[pd.DataFrame, List[Tuple[int,int]]]:
    X_rows = []
    spans = []
    for s, e in sliding_windows(len(df), WIN, STEP):
        ax = df["ax"].values[s:e]
        ay = df["ay"].values[s:e]
        az = df["az"].values[s:e]
        feats = extract_feature_vector(ax, ay, az, fs)
        X_rows.append(feats)
        spans.append((s, e))
    X = pd.DataFrame(X_rows)
    return X, spans

def load_user_dataset(root: str) -> Tuple[pd.DataFrame, np.ndarray]:
    X_list, y_list = [], []
    for folder, label in FOLDER_TO_LABEL.items():
        globpat = os.path.join(root, folder, "*.csv")
        for p in sorted(glob.glob(globpat)):
            df = read_user_csv(p)
            df = trim_and_downsample(df, assume_fs=100, target_fs=50, seconds=FIRST_N_SECONDS)
            Xi, _ = window_and_featurise_user(df, TARGET_FS)
            X_list.append(Xi)
            y_list.extend([label] * len(Xi))
    if not X_list:
        raise RuntimeError(f"No CSV files found under {root}")
    X_user = pd.concat(X_list, axis=0, ignore_index=True)
    y_user = np.array(y_list, dtype=int)
    return X_user, y_user

# ----------------------------
# Train DT on UCI-HAR; Predict on YOUR data
# ----------------------------
def train_dt_on_uci(uci_root: str) -> Tuple[Pipeline, List[str]]:
    ax_tr, ay_tr, az_tr, y_tr = load_uci_split(uci_root, "train")
    ax_te, ay_te, az_te, y_te = load_uci_split(uci_root, "test")

    Xtr = featurise_uci(ax_tr, ay_tr, az_tr, TARGET_FS)
    Xte = featurise_uci(ax_te, ay_te, az_te, TARGET_FS)

    # Same feature columns for all datasets
    feature_cols = list(Xtr.columns)

    clf = Pipeline([
        ("scaler", StandardScaler(with_mean=True, with_std=True)),
        ("dt", DecisionTreeClassifier(
            criterion="gini", max_depth=None, min_samples_leaf=5, random_state=42
        ))
    ])
    clf.fit(Xtr[feature_cols].values, y_tr)

    # Optional sanity check on UCI test
    yte_pred = clf.predict(Xte[feature_cols].values)
    acc = accuracy_score(y_te, yte_pred)
    prec = precision_score(y_te, yte_pred, average="weighted", zero_division=0)
    rec = recall_score(y_te, yte_pred, average="weighted", zero_division=0)
    print(f"[UCI-HAR] Test Acc: {acc:.3f} | Prec: {prec:.3f} | Rec: {rec:.3f}")

    return clf, feature_cols

def evaluate_on_user(clf: Pipeline, feature_cols: List[str], X_user: pd.DataFrame, y_user: np.ndarray):
    # align columns (user may miss some cols if NaNs happen; fill safely)
    for c in feature_cols:
        if c not in X_user.columns:
            X_user[c] = np.nan
    X_user = X_user[feature_cols].copy()
    # fill any remaining NaNs that might appear in short windows
    X_user = X_user.fillna(0.0)

    y_pred = clf.predict(X_user.values)

    acc = accuracy_score(y_user, y_pred)
    prec = precision_score(y_user, y_pred, average="weighted", zero_division=0)
    rec = recall_score(y_user, y_pred, average="weighted", zero_division=0)
    cm = confusion_matrix(y_user, y_pred, labels=sorted(LABEL_TO_NAME.keys()))

    print("\n[YOUR DATA] Metrics")
    print(f"Accuracy : {acc:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"Recall   : {rec:.3f}")
    print("\nClassification report:")
    print(classification_report(y_user, y_pred, target_names=[LABEL_TO_NAME[i] for i in sorted(set(y_user))], zero_division=0))

    # Plot confusion matrix (only classes present in either y or preds for readability)
    present = sorted(set(list(y_user) + list(y_pred)))
    cm_small = confusion_matrix(y_user, y_pred, labels=present)
    names = [LABEL_TO_NAME[i] for i in present]

    fig = plt.figure(figsize=(6, 5))
    plt.imshow(cm_small, interpolation='nearest')
    plt.title("Confusion Matrix - Your Data")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.xticks(range(len(names)), names, rotation=45, ha="right")
    plt.yticks(range(len(names)), names)
    for i, j in product(range(cm_small.shape[0]), range(cm_small.shape[1])):
        plt.text(j, i, cm_small[i, j], ha="center", va="center")
    plt.tight_layout()
    out = "confusion_matrix_user.png"
    plt.savefig(out, dpi=150)
    print(f"\nSaved confusion matrix figure → {out}")

def main():
    # 1) Train on UCI-HAR raw → features
    print("Training Decision Tree on UCI-HAR (raw → features)...")
    clf, feature_cols = train_dt_on_uci(UCI_HAR_ROOT)

    # 2) Load your CSVs, process, window, featurise
    print("Processing your recordings...")
    X_user, y_user = load_user_dataset(USER_DATA_ROOT)
    print(f"Your dataset windows: {X_user.shape[0]} | Features per window: {X_user.shape[1]}")

    # 3) Evaluate on your data
    evaluate_on_user(clf, feature_cols, X_user, y_user)

if __name__ == "__main__":
    main()
```

```python
"""
HAR pipeline (uses ONLY your recorded dataset)
- Reads CSVs from folders (each folder = one activity class)
- Trims each recording to first 10s, downsamples 100Hz -> 50Hz (simple decimation)
- Windows data into 2.56s windows (128 samples @50Hz) with 50% overlap
- Extracts a compact set of time+freq features per window
- Trains a DecisionTreeClassifier on your collected data (train/test split)
- Prints accuracy, precision, recall and confusion matrix and saves the matrix figure

Set USER_DATA_ROOT to the path where your extracted zip lives.
"""

import os
import glob
from pathlib import Path
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis, entropy
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# ========== USER: set this to your extracted folder ==========
USER_DATA_ROOT = r"C:\Users\RAJDEEP VRAJ\OneDrive - iitgn.ac.in\Semester 7\Machine learning\es335-25-fall-assignment-1-master\HAR_task_3"# =============================================================

# Windowing config (matches UCI-HAR)
TARGET_FS = 50             # Hz we want
WIN_SECONDS = 2.56         # 2.56s windows
WIN = int(WIN_SECONDS * TARGET_FS)  # 128 samples
STEP = WIN // 2            # 50% overlap = 64

# Label mapping: folder name (as in your zip) -> integer label
# UCI-style mapping used here (optional). You can change labels if needed.
FOLDER_TO_LABEL = {
    "Upstairs": 2,
    "Downstarirs": 3,   # spelled "Downstarirs" in your archive
    "Sitting": 4,
    "Standing": 5,
    "Laying": 6
}
LABEL_TO_NAME = {v: k for k, v in FOLDER_TO_LABEL.items()}

# ---------- feature extraction helpers ----------
def safe_std(x):
    return float(np.std(x, ddof=1)) if x.size > 1 else 0.0

def median_abs_dev(x):
    return float(np.median(np.abs(x - np.median(x))))

def iqr(x):
    q75, q25 = np.percentile(x, [75, 25])
    return float(q75 - q25)

def energy(x):
    return float(np.sum(x**2) / x.size) if x.size > 0 else 0.0

def zero_crossing_rate(x):
    if x.size < 2: return 0.0
    return float(np.mean(np.diff(np.sign(x)) != 0))

def mean_crossing_rate(x):
    if x.size < 2: return 0.0
    m = np.mean(x)
    return float(np.mean((x[:-1] - m) * (x[1:] - m) < 0))

def spec_features(x, fs):
    """Return spectral centroid, spectral entropy, dominant freq, dominant amplitude"""
    x = np.asarray(x, dtype=float)
    n = x.size
    if n == 0:
        return {"spec_centroid":0.0, "spec_entropy":0.0, "dom_freq":0.0, "dom_amp":0.0}
    # window and rfft
    w = x * np.hanning(n)
    X = np.fft.rfft(w)
    mag = np.abs(X)
    p = mag**2
    freqs = np.fft.rfftfreq(n, d=1.0/fs)
    psum = np.sum(p)
    if psum <= 1e-12:
        return {"spec_centroid":0.0, "spec_entropy":0.0, "dom_freq":0.0, "dom_amp":0.0}
    p_norm = p / psum
    centroid = float(np.sum(freqs * p_norm))
    spec_ent = float(entropy(p_norm + 1e-12, base=2))
    idx = int(np.argmax(mag))
    dom_freq = float(freqs[idx])
    dom_amp = float(mag[idx])
    return {"spec_centroid":centroid, "spec_entropy":spec_ent, "dom_freq":dom_freq, "dom_amp":dom_amp}

def corrcoef_safe(a, b):
    a = np.asarray(a); b = np.asarray(b)
    if a.size < 2 or b.size < 2: return 0.0
    if np.std(a) < 1e-12 or np.std(b) < 1e-12: return 0.0
    return float(np.corrcoef(a, b)[0,1])

def extract_features_from_window(ax, ay, az, fs=TARGET_FS):
    """Return feature dict for one window (ax/ay/az are 1D numpy arrays)"""
    feats = {}
    for name, arr in (("ax", ax), ("ay", ay), ("az", az)):
        arr = np.asarray(arr, dtype=float)
        feats[f"{name}_mean"] = float(np.mean(arr)) if arr.size>0 else 0.0
        feats[f"{name}_std"] = safe_std(arr)
        feats[f"{name}_min"] = float(np.min(arr)) if arr.size>0 else 0.0
        feats[f"{name}_max"] = float(np.max(arr)) if arr.size>0 else 0.0
        feats[f"{name}_median"] = float(np.median(arr)) if arr.size>0 else 0.0
        feats[f"{name}_mad"] = median_abs_dev(arr)
        feats[f"{name}_iqr"] = iqr(arr)
        feats[f"{name}_energy"] = energy(arr)
        feats[f"{name}_zcr"] = zero_crossing_rate(arr)
        feats[f"{name}_mcr"] = mean_crossing_rate(arr)
        feats[f"{name}_skew"] = float(skew(arr)) if arr.size>0 else 0.0
        feats[f"{name}_kurt"] = float(kurtosis(arr, fisher=True)) if arr.size>0 else 0.0
        # spectral
        sf = spec_features(arr, fs)
        feats[f"{name}_spec_centroid"] = sf["spec_centroid"]
        feats[f"{name}_spec_entropy"] = sf["spec_entropy"]
        feats[f"{name}_dom_freq"] = sf["dom_freq"]
        feats[f"{name}_dom_amp"] = sf["dom_amp"]

    # magnitude channel
    m = np.sqrt(np.asarray(ax)**2 + np.asarray(ay)**2 + np.asarray(az)**2)
    feats["mag_mean"] = float(np.mean(m)) if m.size>0 else 0.0
    feats["mag_std"] = safe_std(m)
    feats["mag_energy"] = energy(m)
    sfm = spec_features(m, fs)
    feats["mag_spec_centroid"] = sfm["spec_centroid"]
    feats["mag_spec_entropy"] = sfm["spec_entropy"]
    # correlations
    feats["corr_xy"] = corrcoef_safe(ax, ay)
    feats["corr_xz"] = corrcoef_safe(ax, az)
    feats["corr_yz"] = corrcoef_safe(ay, az)
    return feats

# ---------- helper: sliding windows ----------
def sliding_windows(n_samples, win=WIN, step=STEP):
    start = 0
    while start + win <= n_samples:
        yield start, start + win
        start += step

# ---------- read single CSV (standardize column names) ----------
def read_user_csv(path):
    df = pd.read_csv(path)
    # known column names in your files: 'time','ax (m/s^2)','ay (m/s^2)','az (m/s^2)','aT (m/s^2)'
    colmap = {}
    for c in df.columns:
        cl = c.strip().lower()
        if "time" in cl:
            colmap[c] = "time"
        elif "ax" in cl and "m/s" in cl:
            colmap[c] = "ax"
        elif cl == "ax":
            colmap[c] = "ax"
        elif "ay" in cl and "m/s" in cl:
            colmap[c] = "ay"
        elif cl == "ay":
            colmap[c] = "ay"
        elif "az" in cl and "m/s" in cl:
            colmap[c] = "az"
        elif cl == "az":
            colmap[c] = "az"
        elif "at" in cl or "aT" in c:
            colmap[c] = "aT"
    df = df.rename(columns=colmap)
    # keep at least time, ax, ay, az
    if not {"ax","ay","az"}.issubset(set(df.columns)):
        raise RuntimeError(f"File {path} missing one of ax/ay/az columns. Found: {df.columns.tolist()}")
    # ensure time exists
    if "time" not in df.columns:
        df = df.reset_index().rename(columns={"index":"time"})
        df["time"] = df["time"] / 100.0  # assume 100Hz if no time col
    return df[["time","ax","ay","az"]].copy()

# ---------- trim to first 10 seconds + downsample 100->50 ----------
def trim_downsample(df, orig_fs=100, target_fs=50, seconds=10):
    # keep first `seconds` seconds using time if present
    if "time" in df.columns:
        df_10 = df[df["time"] <= seconds].copy()
        if df_10.empty:
            # fallback: first seconds * orig_fs rows
            nrows = int(seconds * orig_fs)
            df_10 = df.iloc[:nrows].copy()
            df_10["time"] = np.arange(len(df_10)) / orig_fs
    else:
        nrows = int(seconds * orig_fs)
        df_10 = df.iloc[:nrows].copy()
        df_10["time"] = np.arange(len(df_10)) / orig_fs

    factor = int(orig_fs // target_fs)
    if factor <= 0:
        raise ValueError("orig_fs must be >= target_fs and their ratio integer")
    df_ds = df_10.iloc[::factor].reset_index(drop=True)
    # set exact time grid for consistency
    df_ds["time"] = np.arange(len(df_ds)) / target_fs
    # make sure length is exactly seconds * target_fs if possible
    expected = int(seconds * target_fs)
    if len(df_ds) >= expected:
        df_ds = df_ds.iloc[:expected].reset_index(drop=True)
    return df_ds

# ---------- load all user data, featurise windows ----------
def build_feature_matrix(user_root):
    X_rows = []
    y_rows = []
    file_count = 0
    for folder, label in FOLDER_TO_LABEL.items():
        folder_path = os.path.join(user_root, folder)
        if not os.path.isdir(folder_path):
            print(f"Warning: folder not found: {folder_path} (skipping)")
            continue
        csvs = sorted(glob.glob(os.path.join(folder_path, "*.csv")))
        if not csvs:
            print(f"No CSVs found in {folder_path}")
            continue
        for p in csvs:
            file_count += 1
            df = read_user_csv(p)
            df_ds = trim_downsample(df, orig_fs=100, target_fs=TARGET_FS, seconds=10)
            n = len(df_ds)
            if n < WIN:
                # not enough samples to form one window
                print(f"File {p} too short after trimming (n={n}) -> skipping")
                continue
            # form windows
            for s,e in sliding_windows(n, WIN, STEP):
                ax = df_ds["ax"].values[s:e]
                ay = df_ds["ay"].values[s:e]
                az = df_ds["az"].values[s:e]
                feats = extract_features_from_window(ax, ay, az, fs=TARGET_FS)
                X_rows.append(feats)
                y_rows.append(label)
    if file_count == 0:
        raise RuntimeError(f"No files processed under {user_root}")
    X = pd.DataFrame(X_rows)
    y = np.array(y_rows, dtype=int)
    return X, y

# ---------- main flow ----------
def main():
    print("Building feature matrix from user data...")
    X, y = build_feature_matrix(USER_DATA_ROOT)
    print(f"Total windows: {X.shape[0]}, features: {X.shape[1]}")
    print("Class distribution:", dict(zip(*np.unique(y, return_counts=True))))

    # handle any NaNs (rare): fill with 0 or column mean
    if X.isnull().any().any():
        X = X.fillna(0.0)

    # train/test split (stratified)
    splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.30, random_state=42)
    train_idx, test_idx = next(splitter.split(X, y))
    X_train, X_test = X.iloc[train_idx].values, X.iloc[test_idx].values
    y_train, y_test = y[train_idx], y[test_idx]

    # scale features
    scaler = StandardScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_test_s = scaler.transform(X_test)

    # Decision Tree classifier
    clf = DecisionTreeClassifier(criterion="gini", min_samples_leaf=3, random_state=42)
    clf.fit(X_train_s, y_train)

    # predict
    y_pred = clf.predict(X_test_s)

    # metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    rec = recall_score(y_test, y_pred, average="weighted", zero_division=0)
    print("\n=== Results on your collected data (test split) ===")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print("\nClassification report:")
    # map labels present in y_test to names for readable report
    present_labels = sorted(np.unique(y_test))
    target_names = [LABEL_TO_NAME.get(l, str(l)) for l in present_labels]
    print(classification_report(y_test, y_pred, labels=present_labels, target_names=target_names, zero_division=0))

    # confusion matrix
    cm = confusion_matrix(y_test, y_pred, labels=present_labels)
    fig, ax = plt.subplots(figsize=(6, 5))
    im = ax.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
    ax.set_title("Confusion Matrix (your data)")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")
    ax.set_xticks(np.arange(len(present_labels)))
    ax.set_yticks(np.arange(len(present_labels)))
    ax.set_xticklabels(target_names, rotation=45, ha="right")
    ax.set_yticklabels(target_names)
    # annotate numbers
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, int(cm[i, j]), ha="center", va="center", color="black")
    fig.colorbar(im, ax=ax)
    plt.tight_layout()
    out_fig = "confusion_matrix_user_data.png"
    fig.savefig(out_fig, dpi=150)
    print(f"Saved confusion matrix to: {out_fig}")

if __name__ == "__main__":
    main()
```
